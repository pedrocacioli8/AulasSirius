```python
def  dens_dem(df, coluna, limite):
  cat = []
  for valor in df[coluna]:
    if valor <= limite:
      cat.append(0)
    else:
      cat.append(1)

  return cat
```

```python
df["Densidade demográfica categórica"] = dens_dem(df, "Densidade demográfica", 100)
df.head()
```

```python
def reordena_colunas(df, nome_coluna, index):
  cols = df.columns.tolist()
  cols.remove(nome_coluna)
  cols.insert(index, nome_coluna)

  return df[cols]
```

```python
df = reordena_colunas(df, "Densidade demográfica categórica", 4)
df.head()
```
