```python
import pandas as pd
pd.set_option('display.max_columns', None)
df = pd.read_excel('Fonte dos Dados (1).xlsx')
df.head()
```

```python
df = df.iloc[:,3:5]
```

```python
df.info()
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=2)
kmeans.fit(df)
```

```python
kmeans = KMeans(n_clusters=3).fit(df)
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
kmeans.labels_
```

```python
df_clusterizado = df
```

```python
df_clusterizado['label']=kmeans.labels_
df_clusterizado
```

```python
df_clusterizado[df_clusterizado['label']==1]
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
plt.scatter (df['Densidade demográfica'], df['Porcentagem de Jovens'], c=kmeans.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:,0], centroids[:,1], c='red', s=50)
plt.show()
```
