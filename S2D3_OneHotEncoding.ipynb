```python
import pandas as pd
```

```python
consumidores = pd.read_csv("customer_supermarket.xlsx - customer_supermarket.csv")
```

```python
consumidores.head()
```

```python
consumidores.dropna(inplace=True)
```

```python
consumidores = consumidores[["Income", "Education"]]
consumidores.head()
```

```python
consumidores["Education"].unique()
```

```python
consumidores["Education"] = consumidores["Education"].str.replace("Basic", "Ensino_Basico")
consumidores["Education"] = consumidores["Education"].str.replace("Graduation", "Ensino_Superior")
consumidores["Education"] = consumidores["Education"].str.replace("PhD", "PosGrad")
consumidores["Education"] = consumidores["Education"].str.replace("Master", "PosGrad")
consumidores["Education"] = consumidores["Education"].str.replace("2n Cycle", "PosGrad")
```

```python
consumidores["Education"].value_counts()
```

```python
pd.get_dummies(consumidores, columns=["Education"])
```

```python
consumidores = pd.get_dummies(consumidores, columns=["Education"], prefix=" ")
consumidores.head()
```

```python
consumidores.drop("_Ensino_Basico", axis=1, inplace=True)
consumidores.head()
```
