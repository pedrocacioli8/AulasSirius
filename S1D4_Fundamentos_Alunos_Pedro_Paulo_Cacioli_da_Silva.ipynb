### No entanto veja que:
Para interpretar os resultados da função corr(), é importante considerar o intervalo de valores possíveis para os coeficientes de correlação. Como mencionado anteriormente, os coeficientes de correlação variam de -1 a 1.

    Um coeficiente de correlação de 1 indica uma forte relação positiva entre as duas variáveis. Isso significa que, quando uma variável aumenta, a outra variável também aumenta.
    Um coeficiente de correlação de -1 indica uma forte relação negativa entre as duas variáveis. Isso significa que, quando uma variável aumenta, a outra variável diminui.
    Um coeficiente de correlação próximo de 0 indica que não há relação entre as duas variáveis.

É importante notar que a correlação não implica causalidade. Apenas porque duas variáveis estão correlacionadas não significa que uma causa a outra. Por exemplo, o fato de que as vendas de sorvetes aumentam no verão não significa que o sorvete cause o verão.

h) Criar uma coluna que mostre a classificação da obesidade com base na tabela de IMC abaixo

image.png

```python
# Criar uma nova coluna com a coluna classificação de magreza
# Se necessário, crie uma função para aplicar os valores.

df.loc[(df['BMI'] < 18.5), 'ClassificacaoMagreza'] = 'MAGREZA'
df.loc[(df['BMI'] >= 18.5) & (df['BMI'] < 25), 'ClassificacaoMagreza'] = 'NORMAL'
df.loc[(df['BMI'] >= 25) & (df['BMI'] < 30), 'ClassificacaoMagreza'] = 'SOBREPESO'
df.loc[(df['BMI'] >= 30) & (df['BMI'] < 40), 'ClassificacaoMagreza'] = 'OBESIDADE'
df.loc[(df['BMI'] >= 40), 'ClassificacaoMagreza'] = 'OBESIDADE GRAVE'

df.head()
```
